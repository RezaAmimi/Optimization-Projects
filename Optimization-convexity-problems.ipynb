$$f_1(x) := x^{4}-x^{3}-x^{2}, \ \ x\in[-3,3]$$
$$f_2(x) := \frac{1}{2}x^{2}\sin\left(2x\right) , \ x\in [-3,3]$$

$$f_4(x) := \left( 1.5 - x_1 + x_1x_2 \right)^{2} + \left( 2.25 - x_1 + x_1x_2^{2}\right)^{2}+ \left( 2.625 - x_1 + x_1x_2^{3}\right)^{2},  \ \ x=(x_1,x_2)\in [-4,4]^2.$$

    Verify if $f_1, f_2, f_3$, and $f_4$ are convex or not. To prove that a function $f$ is convex, you need to show that for *every* $x^1$ and $x^2$ in the domain of the function and *every* $\beta\in[0,1]$, one has
    $$f(\beta x^1 + (1-\beta)x^2)\le \beta f(x^1) + (1-\beta) f(x^2).$$ 
    To show that $f$ is *not* convex, it suffices to find two points $\hat x^1$ and $\hat x^2$ in the domain of the function as well as a constant $\hat \beta\in[0,1]$ such that the inequality
    $$f(\hat\beta \hat x^1 + (1-\hat \beta)\hat x^2) >  \hat \beta f(\hat x^1) + (1-\hat \beta) f(\hat x^2),$$
    holds. Note that if $f$ is 1D, then $x^1, x^2, \hat x^1,$ and $\hat x^2$ in above expressions are numbers, but if $f$ is 2D, then these are 2D vectors. <br></br>

````python
"""
____________________________
Code for Question 1. Part 2.
‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾
"""

from sympy import symbols, simplify, expand
import sympy
import itertools

# For f1
x = np.linspace(-0.8, 0.5, 100)
plt.plot(x, f1(x), [-0.76, 0.29], [0.2, -0.1])
plt.title("f1")
plt.show()

# For f2
x = np.linspace(-1, 1, 100)
plt.plot(x, f2(x), [-0.75, 0.75], [-0.27, 0.28])
plt.title("f2")
plt.show()


# For f3

sympy.init_printing()

x = [sympy.Symbol('x_'+str(i+1)) for i in range(2)]
alpha = sympy.Symbol('a')


def f3p(x): 
    return x[0] + x[1] - x[0]**0.5*x[1]**(1 - 0.5)

SymFn = f3p(x)

xbar = [0,1]
d = [2,1]

fd = SymFn.subs([(x[0],xbar[0]+d[0]*alpha),(x[1],xbar[1]+d[1]*alpha)])

fdeval = sympy.lambdify([alpha],fd)

a = np.linspace(0,1,1000)
y = fdeval(a)
plt.plot(a,fdeval(a), [0.02,0.22], [0.85,0.93])

plt.xlabel(r"$\alpha$",fontsize=16)
plt.ylabel(r"$g(\alpha)$",fontsize=16)

plt.show()

fdDer = fd.diff(alpha)
fdDer2 = fdDer.diff(alpha)
fdDer.subs(alpha,0)
fdDer2.subs(alpha,0)


# For f4
def f4p(x): 
    return (1.5 - x[0] + x[0]*x[1])**2 + (2.25  - x[0] + x[0]*x[1]**2)**2 + (2.625 - x[0] + x[0]*x[1]**3)**2

SymFn = f4p(x)

xbar = [3,0]
d = [-1,-1]

fd = SymFn.subs([(x[0],xbar[0]+d[0]*alpha),(x[1],xbar[1]+d[1]*alpha)])

fdeval = sympy.lambdify([alpha],fd)

a = np.linspace(-0.9,0.9,1000)
y = fdeval(a)
plt.plot(a,fdeval(a), [-0.5,0.5], [0.5,5])

plt.xlabel(r"$\alpha$",fontsize=16)
plt.ylabel(r"$g(\alpha)$",fontsize=16)

plt.show()

fdDer = fd.diff(alpha)
fdDer2 = fdDer.diff(alpha)
fdDer.subs(alpha,0)
fdDer2.subs(alpha,0)


#Hessian for f4 
alpha = 0.2
x, x1, x2, beta, a, b, b1, b2 = symbols("x, x1, x2, beta, a, b, b1, b2")
f3s = x1 + x2 - x1**alpha*x2**(1 - alpha)
H = sympy.hessian(f3s, [x1, x2])
sympy.pprint(H)
B = sympy.Matrix([b1, b2]);
K = B.T*H*B
sympy.simplify(K[0])
````

In this question, you will learn how to apply golden section search (GSS; see Sections 3.3 and 3.4 of [Kochenderfer and Wheeler (2019)](#Kochenderfer-and-Wheeler-2019)) for minimizing a univariate function $f(x):[a,b]\mapsto\mathbb{R}$. Recall functions $f_1$ and $f_2$ from Question 2.

Find the minimum of $f_1$ and $f_2$ using GSS starting from the initial bracket $[a,b]=[-3,3]$. Perform four iterations of the search and report the solution after each iteration. Specifically, create a table that shows the bracket computed from GSS at each iteration for both functions. 

### **Answer.**

Golden-Section Search function is written and the extermum of f1 and f2 function in the given range is obtained. In each iteration, error and current extermum is printed.

````python

def GSS(f, x1, x2):
    
    golden_ratio = (5**0.5 + 1) / 2

    x3 = x2 - (x2 -x1)/golden_ratio
    x4 = x1 + (x2 - x1)/golden_ratio
    while abs(x2 - x1) > 1e-5:
        if f(x3) < f(x4):
            x2 = x4
        else:
            x1 = x3

        x3 = x2 - (x2 - x1)/golden_ratio
        x4 = x1 + (x2 - x1)/golden_ratio
        
        print("Error: ", abs(x2 - x1), ", Extermum: ", (x1 + x2)/2)

    return (x2 + x1)/2

print("F1 function:")
print(GSS(f1, -3, 3))
print("--------------------")
print("F2 function:")
print(GSS(f2, -3, 3))
````

$$\mathrm{OLS}(\beta; \mathcal{T}) :=  \frac{-x_i}{10‍11}\sum_{i=1}^{1022} (y_i - \beta^\top x_i) $$
$$\mathrm{Ridge}(\beta; \alpha,\mathcal{T}) :=  \frac{-x_i}{1011}\sum_{i=1}^{1022} (y_i - \beta^\top x_i) + \frac{\alpha \beta^\top}{\lVert \beta\rVert_2} $$

````python
""" ___________
    Sample Code
    ‾‾‾‾‾‾‾‾‾‾‾
"""

import numpy as np
import pandas as pd
from autograd import grad
import matplotlib.pyplot as plt

class Regression():

    def __init__(self,train_X,train_y) -> None:
        # Add description, input parameters, and output of this function here
        self.train_X = train_X
        self.train_y = train_y
        

    def OLS(self,beta:np.ndarray)->float:
        """
        Description
        ‾‾‾‾‾‾‾‾‾‾‾    
        This function returns the value of 2-norm at a given weight vector beta.

        Parameters
        ‾‾‾‾‾‾‾‾‾‾
        - beta [np.ndarray]: value of weight vector \beta.

        Returns  
        ‾‾‾‾‾‾‾
        - returns the value of OLS function in Part 1 for  weight vector beta and the train dataset given by (self.train_y,self.train_X)
        """
        sum2 = 0.0
        for i in range(len(self.train_y)):
            sum1 = 0.0
            for j in range(len(self.train_X.columns)):
                sum1 = sum1 + beta[j]*self.train_X.iloc[i,j]
            sum2 = sum2 + (y[i] - sum1)**2
        return 1.0/1022.0*sum2
    
    def Ridge(self,beta:np.ndarray, alpha:float)->float:
        """
        Description
        ‾‾‾‾‾‾‾‾‾‾‾    
        This function returns the value of 2-norm at a given weight vector beta.

        Parameters
        ‾‾‾‾‾‾‾‾‾‾
        - beta [np.ndarray]: value of weight vector \beta.

        Returns  
        ‾‾‾‾‾‾‾
        - returns the value of OLS function in Part 1 for  weight vector beta and the train dataset given by (self.train_y,self.train_X)
        """

        sum2 = 0.0
        norm_beta = 0.0
        for i in range(len(beta)):
            norm_beta = norm_beta + beta[i]**2
        norm_beta = norm_beta**0.5    
        for i in range(len(self.train_y)):
            sum1 = 0.0
            for j in range(len(self.train_X.columns)):
                sum1 = sum1 + beta[j]*self.train_X.iloc[i,j]
            sum2 = sum2 + (y[i] - sum1)**2 + alpha*norm_beta
        return 1.0/1022.0*sum2
    

    def OLS_gradient(self, beta:np.ndarray)->np.ndarray:
        """
        Description
        ‾‾‾‾‾‾‾‾‾‾‾    
        This function returns the gradient of 2-norm at a given weight vector beta and a datapoint

        Parameters
        ‾‾‾‾‾‾‾‾‾‾
        - beta [np.ndarray]: value of weight vector \beta.

        Returns  
        ‾‾‾‾‾‾‾
        - returns the gradient of OLS function in Part 1 for weight vector beta and the train dataset given by (self.train_y,self.train_X)
        """
        OLS_grad = grad(self.OLS)
        return OLS_grad(beta)
    
    def Ridge_gradient(self, beta:np.ndarray, alpha:float)->np.ndarray:
        """
        Description
        ‾‾‾‾‾‾‾‾‾‾‾    
        This function returns the gradient of 2-norm at a given weight vector beta and a datapoint

        Parameters
        ‾‾‾‾‾‾‾‾‾‾
        - beta [np.ndarray]: value of weight vector \beta.

        Returns  
        ‾‾‾‾‾‾‾
        - returns the gradient of OLS function in Part 1 for weight vector beta and the train dataset given by (self.train_y,self.train_X)
        """
        Ridge_grad = grad(self.Ridge)
        return Ridge_grad(beta, alpha)
    

if __name__== "__main__":

    """
    Load data
    """
    train = pd.read_csv('home_price_clean_train.csv')
    test  = pd.read_csv('home_price_clean_test.csv')
    print(train.describe())
    print(test.describe())
    fig, ax  = plt.subplots(figsize=(6, 6),dpi=100)
    ax.matshow(train.corr())

    """
    Part (1)
    """
    y = train['SalePrice']
    X = train
    del X['SalePrice'] 

    Reg = Regression(X,y)
    OLS_value = Reg.OLS(np.ones((len(X.columns), 1)))
    OLD_grad_value = Reg.OLS_gradient(np.ones((len(X.columns), 1)))

    Ridge_value = Reg.Ridge(np.ones((len(X.columns), 1)), 1)
    Ridge_grad_value = Reg.Ridge_gradient(np.ones((len(X.columns), 1)), 1)
    
    print("OLS function value at (1,1, .., 1) = ", OLS_value)
    print("OLS gradient function value at (1,1, .., 1) = ", OLD_grad_value)
    print("Ridge function value at (1,1, .., 1) = ", Ridge_value)
    print("Ridge gradient function value at (1,1, .., 1) = ", Ridge_grad_value)
````
